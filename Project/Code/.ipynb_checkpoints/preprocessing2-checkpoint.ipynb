```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn

training = pd.read_csv("../Data/train_data_with_headers.csv")
testing = pd.read_csv("../Data/test_data_with_headers.csv")

x_training = training.iloc[:,0:152] # the dataset without the last column (the prediction target)
y_training = training.iloc[:,152]

x_testing = testing.iloc[:,0:152]
y_testing = testing.iloc[:,152]

x_training.head()
```

### Step 1: Find columns whose results are all 0 or 1, and remove them from the dataframe

```python
training = training.loc[:,(training!=0).any()]
training = training.loc[:,(training!=1).any()]

print(training.shape)
training.head()
```

### Step 2: Find duplicate columns and remove one of them from the dataframe
We can remove one of columns from each of the pairs below because they are duplicates of each other:
* column"frame.time_delta" and "frame.time_delta_displayed"
* column "frame.len" and "frame.cap_len"
* column "wlan.qos.tid" and "wlan.qos.priority"

```python
print(sum(training["frame.time_delta"] == training["frame.time_delta_displayed"]) == training.shape[0])
print(sum(training["frame.len"] == training["frame.cap_len"]) == training.shape[0])
print(sum(training["wlan.qos.tid"]==training["wlan.qos.priority"]) == training.shape[0])

training = training.drop(["frame.time_delta_displayed","frame.cap_len","wlan.qos.priority"],axis=1)
training.shape
```

### Step 3. Plot distribution of values in each column in relation to the prediction result
View all columns and select those where the distribution appears different for the two classes

```python
# use two-sided Kolmogorov-Smirnov test to find statistical difference between 2 distributions
from scipy.stats import ks_2samp 
col_pvalues = {} 

# remove columns whose p value is > 0.05
for i, column in enumerate(training.columns):
    null_class = training[column].where(training["class"]==0).dropna()
    yes_class = training[column].where(training["class"]==1).dropna()
    col_pvalues[column]= ks_2samp(null_class,yes_class)[1]

remove_columns = list({k: v for k, v in col_pvalues.items() if v>0.05}.keys())
training = training.drop(remove_columns,axis=1)

# plot the distribution of each column with regards to the prediction target
fig=plt.figure(figsize=(20,40))

for i, column in enumerate(training.columns):
    ax=fig.add_subplot(11,4,i+1)
    null_class = training[column].where(training["class"]==0).dropna()
    yes_class = training[column].where(training["class"]==1).dropna()
    plt.title(column+" Distribution")
    plt.hist(null_class, alpha=0.5, label='null')
    plt.hist(yes_class, alpha=0.5, label='yes')
    plt.legend(loc='upper right')

fig.tight_layout()
plt.show()
```

```python
# Looking at the charts above, we can see that the following columns can also be removed since their distribution value 
col_manual_removal = ["frame.time_delta",
             "wlan.ra",
             "wlan_mgt.fixed.capabilities.ess",
             "wlan_mgt.fixed.capabilities.privacy",
             "wlan_mgt.fixed.capabilities.short_slot_time",
             "wlan_mgt.fixed.auth_seq",
             "wlan.tkip.extiv"
            ]

training = training.drop(col_manual_removal,axis=1)
training.shape
```

### Step 4: Rescale columns by using `MinMaxScaler`

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
X = training.iloc[:,0:36].values
rescaledX = scaler.fit_transform(X)
training = pd.DataFrame(rescaledX, columns = list(training.columns.values)[0:36])

X2 = testing.iloc[:,0:36].values
rescaledX2 = scaler.fit_transform(X2)
testing = pd.DataFrame(rescaledX2, columns = list(testing.columns.values)[0:36])
```

### Step 5: Normalise columns by using `Normalizer`

```python
from sklearn.preprocessing import Normalizer

X = training.iloc[:,0:36].values
normScaler = Normalizer().fit(X)
normalizedX = normScaler.transform(X)
training = pd.DataFrame(normalizedX, columns = list(training.columns.values)[0:36])

X2 = testing.iloc[:,0:36].values
normScaler = Normalizer().fit(X2)
normalizedX2 = normScaler.transform(X2)
testing = pd.DataFrame(normalizedX2, columns = list(testing.columns.values)[0:36])
```

```python
# Plot the final dataset with rescaling
def draw_histograms(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(20,40))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=10,ax=ax)
        ax.set_title(var_name+" Distribution")
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()

draw_histograms(training, training.columns, 9, 4)
```

```python
training =pd.concat([training,y_training],axis=1)
testing = pd.concat([testing,y_testing],axis=1)
```

```python
training.to_csv("../Data/preprocessed_training_with_scaling.csv")
testing.to_csv("../Data/preprocessed_testing_with_scaling.csv")
```
